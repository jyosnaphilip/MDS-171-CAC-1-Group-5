```python
import random
customers={}    #stores customer details in accountnumber:{detailname:details} format
```

```python
f=open("CustomerDetails.csv","w+")
f.write("Account Number,name,age,aadhar,phone number,account type\n")
f.close()
def createAccount():
    accountNum=""
    for i in range(10):                             #generation of acc.number
        j=str(random.randint(0,9))
        accountNum=accountNum+j
    details_keys=["name","age","aadhar","phone number","account type"]                          
    name=input("enter name:")
    age=input("enter age")
    aadhar=input("enter aadhar number")               
    phno=input("enter mobile number")
    accType=input("enter account type: Savings or Current?")
    details_values=[name.title().strip(),age,aadhar,phno,accType.title().strip()]
    details=dict(zip(details_keys,details_values))
    customers.update({accountNum:details})  #details are getting updated, customer list updated with account number and details
    
    with open("CustomerDetails.csv","a+") as file:
        file.write(accountNum+",")
        for i in details_values:
            file.write(str(i))     #writing to file
        file.write("\n")
    print("account created successfully!")
    print("your account Number is ",accountNum)       #writing string to file
```

```python
def editName(line):
    
    
    oldName=(input("enter old name")).strip().title()
    newName=(input("enter new name")).strip().title()
    if oldName in line:
        new_line=line.replace(oldName,newName)
        with open("CustomerDetails.csv","a+") as file:
            file.write(new_line)
        with open("CustomerDetails.csv","r+") as file:
            a=file.readlines()
        
        a.remove(line)
        
        with open("CustomerDetails.csv","a+") as file:
            file.writelines(a)
            
        print("name updated")
    else:
        print("name doesnt match !")
```

```python
def editPhone(line):
    oldph=(input("enter your phone no:"))
    newph=input("enter new phone number:")
    if oldph in line:
        new_line=line.replace(oldph,newph)
        with open("CustomerDetails.csv","a+") as file:
            file.write(new_line)
        with open("CustomerDetails.csv","r+") as file:
            a=file.readlines()
        
        a.remove(line)
        
        with open("CustomerDetails.csv","a+") as file:
            file.writelines(a)
            
        print("phone number updated")
    else:
        print("phone number doesnt match !")
```

```python
def editDetails():
    while True:
        print("a.edit name")
        print("b.edit phone number")
        print("c.exit")
        option=input("enter your option:")
        if option=="a" or option=="b":
            ac=input("enter account number")
            with open("CustomerDetails.csv","r+") as file:
                for line in file.readlines():
                    if ac in line:
                        if option=="a":
                            editName(line)
                        elif option=="b":
                            editPhone(line)
                        else:
                            pass

                else :
                    print("account number not found!")

        elif option=="d":
            break
        else:
            print("enter valid option")
```

```python
def inr_to_usd(amount):
    conversionRate=83.1
    serviceCharge=amount*0.01
    usd=(amount-serviceCharge)/conversionRate
    print("amount recieved\t\t{} INR".format(amount))
    print("\tservice charge:1%' of {} INR".format(amount))
    print("\t\t\t=",serviceCharge)
    print("converted amount:USD ",usd)
    print("\t\tconversion rate=",1/conversionRate)

def usd_to_inr(amount):
    conversionRate=83.1
    serviceCharge=amount*0.01
    rs=(amount-serviceCharge)*conversionRate            
    print("amount recieved\t\t{} USD".format(amount))
    print("\tservice charge:1%' of {} USD".format(amount))
    print("\t\t\t=",serviceCharge)
    print("converted amount:INR ",rs)
    print("\t\tconversion rate=",conversionRate)
    
```

```python
def inr_to_aed(amount):
    convRate=22.66  
    serviceCharge=amount*0.01
    aed=(amount-serviceCharge)/convRate 
    print("amount recieved\t\t{} INR".format(amount))
    print("\tservice charge:1%' of {} INR".format(amount))
    print("\t\t\t=",serviceCharge)
    print("converted amount:AED ",aed)
    print("\t\tconversion rate=",1/convRate)
def aed_to_inr(amount):
    convRate=22.66  
    serviceCharge=amount*0.01
    aed_to_inr(amount)
    rs=(amount-serviceCharge)*convRate            
    print("amount recieved\t\t{} AED".format(amount))
    print("\tservice charge:1%' of {} AED".format(amount))
    print("\t\t\t=",serviceCharge)
    print("converted amount:INR",rs)
    print("\t\tconversion rate=",convRate)
    
```

```python
def inr_to_euro(amount):
    convRate=88.72
    serviceCharge=amount*0.01
    euro=(amount-serviceCharge)/convRate            
    print("amount recieved\t\t{} INR".format(amount))
    print("\tservice charge:1%' of {} INR".format(amount))
    print("\t\t\t=",serviceCharge)
    print("converted amount:EURO ",euro)
    print("\t\tconversion rate=",1/convRate)
def euro_to_inr(amount):
    convRate=88.72
    serviceCharge=amount*0.01
    rs=(amount-serviceCharge)*convRate            
    print("amount recieved\t\t{} EURO".format(amount))
    print("\tservice charge:1%' of {} EURO".format(amount))
    print("\t\t\t=",serviceCharge)
    print("converted amount:INR",rs)
    print("\t\tconversion rate=",convRate)
    
```
