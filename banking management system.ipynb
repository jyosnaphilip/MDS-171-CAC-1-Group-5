```python
import random
customers={}    #stores customer details in accountnumber:{detailname:details} format
customerTransactions={}  #stores transaction in accountnumber:[transactions] format 
```

```python
f=open("CustomerDetails.csv","w+")
f.write("Account Number,name,age,aadhar,phone number,account type,Transaction details,Balance\n")
f.close()
def createAccount():
    accountNum=""
    for i in range(10):                             #generation of acc.number
        j=str(random.randint(0,9))
        accountNum=accountNum+j
    details_keys=["name","age","aadhar","phone number","account type","Transactions"]                          
    name=input("enter name:")
    age=input("enter age")
    aadhar=input("enter aadhar number")               
    phno=input("enter mobile number")
    accType=input("enter account type: Savings or Current?")
    details_values=[name.title().strip(),age,aadhar,phno,accType.title().strip(),[]]
    details=dict(zip(details_keys,details_values))
    customers.update({accountNum:details})  #details are getting updated, customer list updated with account number and details
    customerTransactions.update({accountNum:[]})     
         
    
    
    with open("CustomerDetails.csv","a+") as file:
        file.write(accountNum+",")
        for i in details_values:
            file.write(str(i))     #writing to file
        file.write(str(customerTransactions[accountNum]))   #also adding list to store transactions
        file.write("\n")
    print("account created successfully!")
    print("your account Number is ",accountNum)       #writing string to file

 

    
```

```python
def editName(line):
    
    
    oldName=(input("enter old name")).strip().title()
    newName=(input("enter new name")).strip().title()
    if oldName in line:
        new_line=line.replace(oldName,newName)
        with open("CustomerDetails.csv","a+") as file:
            file.write(new_line)
        with open("CustomerDetails.csv","r+") as file:
            a=file.readlines()
        
        a.remove(line)
        
        with open("CustomerDetails.csv","a+") as file:
            file.writelines(a)
            
        print("name updated")
    else:
        print("name doesnt match !")
    
    
```

```python
def editPhone(line):
    oldph=(input("enter your phone no:"))
    newph=input("enter new phone number:")
    if oldph in line:
        new_line=line.replace(oldph,newph)
        with open("CustomerDetails.csv","a+") as file:
            file.write(new_line)
        with open("CustomerDetails.csv","r+") as file:
            a=file.readlines()
        
        a.remove(line)
        
        with open("CustomerDetails.csv","a+") as file:
            file.writelines(a)
            
        print("phone number updated")
    else:
        print("phone number doesnt match !")
    
```

```python
def editDetails():
    while True:
        print("a.edit name")
        print("b.edit phone number")
        print("c.exit")
        option=input("enter your option:")
        if option=="a" or option=="b":
            ac=input("enter account number")
            with open("CustomerDetails.csv","r+") as file:
                for line in file.readlines():
                    if ac in line:
                        if option=="a":
                            editName(line)
                        elif option=="b":
                            editPhone(line)
                        else:
                            pass

                else :
                    print("account number not found!")

        elif option=="d":
            break
        else:
            print("enter valid option")


    
```

```python
def loanEligibility():
    print("are you....")
    print("1.self employed")
    print("2.salaried")

    choice=input("enter job option:")
    age=int(input("enter your age:"))
    income=int(input("enter your annual income:"))
    creditScore=int(input("enter your credit score"))
    if choice=="1":
        if creditScore>=750 and 21<=age<=65 and income>=1500000:
            year=int(input("enter period of business coninuity:"))
            if year>=3:
                print("you are eligible for a loan!!")
            else:
                print("not eligible for loan")
        else:
            print("not eligible for loan")
    
    elif choice=="2":
        if 18<=age<=60 and creditScore>=750 and income>=180000:
            year=int(input("enter period of employment in years:"))
            if year>=1:
                print("you are eligible for a loan!")
            else:
                print("not eligible for loan")
        else:
            print("not eligible for loan")
    else:
        print("not a valid option")
                
```
